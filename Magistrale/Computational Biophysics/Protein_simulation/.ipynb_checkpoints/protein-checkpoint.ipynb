```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x,y=np.loadtxt('potential.xvg',comments=["#","@"],unpack=True)
print(x,"\n",y)
```

```python
fig,ax=plt.subplots()
ax.plot(x,-y)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel("t [ns]")
ax.set_ylabel("Energy")
ax.set_title("Energy vs Time")
```
