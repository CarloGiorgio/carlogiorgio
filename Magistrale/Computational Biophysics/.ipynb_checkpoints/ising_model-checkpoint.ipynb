# Ising Model
## Swesden Wang algorithm
We try to simulate with a Swesden-Wang algorithm Ising Model.

In this algorithm we individuate two neighbours site with the same spin and with a probability equal to 
$$1-e^{-2J\beta}$$
We decide to put them in the same cluster.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
L=20 #lattice 
N=L*L #number of spins
T=2.3 #temperature
```

```python
S=2*np.random.randint(2,size=(L,L))-1
plt.imshow(S,'binary')
S=S.reshape(N)
print(S)
```

```python
#neighbours list
node=[[(i+1)%L+i//L*L,
       (i-1)%L+i//L*L,
       (i+L)%N,
       (i-L)%N] for i in range(N)]
```

## Cluster creation
The following function create a cluster. First it check if the element is not in a cluster and then try to put it in one.
The algorithm bust be done by recursion!

```python
def trynna(gi,node,l,k):
    for n in node[gi]:
        if S[n]==S[gi] and n in k and np.random.random()<1-np.exp(-2./T) :
            l.append(n)
            k.remove(n)
            trynna(n,node,l,k)

def clust_creation(nodenext):
    g=list(range(N))
    k=g.copy()
    cl=[]
    j=0
    gi=0
    l=[]
    while True:
        if (gi in k) and j!=0:
            l.append(s)
            cl.append(l.copy())
            s=gi
            l.clear()
        if (gi in k) and j==0:
            s=gi
            j=1
        if gi in k:
            k.remove(gi)
            trynna(gi,nodenext,l,k)
        if not k:
            break
        gi=k[0]
        
    l.append(s)
    cl.append(l.copy())
    return cl.copy()

cl=clust_creation(node)
```

```python
mm=[]
m2=[]
M=[]
MM=[]
t=500
T_r=np.linspace(0.01,3.2)
for T in T_r:
    for _ in range(t):
        cl=clust_creation(node)
        if np.random.random() < 0.5:
                S[cl[np.random.randint(len(cl))]]*=-1
        ms=m(S)
        mm.append(ms)
        m2.append(ms*ms)
    M.append(sum([abs(mm[i]) for i in range(len(mm))])/t)
    MM.append((sum(m2)/t-M[-1]**2))
    mm.clear()
    m2.clear()
```
